### Autocorrelation

```python
weekly_sales_lags = weekly_sales.loc[weekly_sales['store']==1,].copy()
weekly_sales_lags.drop(columns=['store'], inplace=True)
weekly_sales_lags['sales_lag_m'] = weekly_sales_lags['weekly_sales'].shift(periods = 4)
weekly_sales_lags['sales_lag_q'] = weekly_sales_lags['weekly_sales'].shift(periods = 13)
weekly_sales_lags['sales_lag_y'] = weekly_sales_lags['weekly_sales'].shift(periods = 52)

plt.figure(figsize=(10, 10))
mask = np.zeros_like(weekly_sales_lags.corr())
mask[np.triu_indices_from(mask)] = True

sns.heatmap(weekly_sales_lags.corr(),
            mask=mask,
            annot=True,
            cmap='coolwarm',
            vmin=-1, 
            vmax=1);
plt.yticks(rotation=0, fontsize=14)
plt.xticks(fontsize=14)
plt.title('Correlation of Lags, Store ID 1', fontsize=20, fontweight='bold')
;
```

Using one store id to illustrate, only the one-year lag appears highly correlated. This suggests that weekly sales is not stable over time, because so many lagged correlations are low, but that an annual cycle occurs because values for a given week are closely associated with that from the previous year. 

```python
i=1
fig, ax = plt.subplots(5,2,figsize=(14,20))
for i in range(5):
    plot_acf(weekly_sales.loc[weekly_sales['store']==(i+1),'weekly_sales'], lags=52, ax=ax[i,0])
    plot_pacf(weekly_sales.loc[weekly_sales['store']==(i+1),'weekly_sales'], lags=52, ax=ax[i,1])
plt.suptitle('Autocorreltation Plots, First 5 Store IDs', fontsize=22, fontweight='bold')
;
```

Examing the partial correlation plots in the right column, we find a lag of one week is significant across our 5 store IDs. We find inconsistency in significane for later lags, but we find a lag of 4 weeks appears significant for four of the values and a negative correlation at 5 week controlling for those ahead. We will proceed using one lag in our model.

## Stationarity

```python
station = 0
non_station=[]
for sid in range(45):
    print('Store {} - ADFuller Stat: {:.3f}, p-value{:.3f}'.format(sid+1, \
                adfuller(weekly_sales.loc[weekly_sales['store']==(sid+1),['weekly_sales']])[0], \
                adfuller(weekly_sales.loc[weekly_sales['store']==(sid+1),['weekly_sales']])[1]))
    if adfuller(weekly_sales.loc[weekly_sales['store']==(sid+1),['weekly_sales']])[0] < \
        adfuller(weekly_sales.loc[weekly_sales['store']==(sid+1),['weekly_sales']])[4]['5%']:
        station += 1
    else:
        non_station.append(sid+1)

print(' ')
print('{} Stores out of 45 are Stationary at 95% CL'.format(station))
print('Stores with sales which are not stationary: {}'.format(non_station))
```
