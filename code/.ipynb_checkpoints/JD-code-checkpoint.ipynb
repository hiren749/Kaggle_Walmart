#### 3. Autocorrelations

We've learned about autocorrelation: the influence a serial observation has on the observations which follow.

Calculate autocorrelation in your `Weekly_Sales` data so that you can answer the question: 'How does autocorrelation differ as lag-weeks vary?'

**3a. Check at least three lag intervals (but feel free to explore others):**
 - **1 week**
 - **2 weeks**
 - **1 year (52 weeks)**

```python
weekly_sales_lags = weekly_sales.copy()
weekly_sales_lags['sales_lag1'] = weekly_sales_lags['weekly_sales'].shift()
weekly_sales_lags['sales_lag2'] = weekly_sales_lags['sales_lag1'].shift()
weekly_sales_lags['sales_lag3'] = weekly_sales_lags['sales_lag2'].shift()
weekly_sales_lags['sales_lag4'] = weekly_sales_lags['sales_lag3'].shift()

plt.figure(figsize=(8, 8))
mask = np.zeros_like(weekly_sales_lags.corr())
mask[np.triu_indices_from(mask)] = True

sns.heatmap(weekly_sales_lags.corr(),
            mask=mask,
            annot=True,
            cmap='coolwarm',
            vmin=-1, 
            vmax=1);
```

```python
weekly_sales_lags = weekly_sales.copy()
weekly_sales_lags['sales_lag_m'] = weekly_sales_lags['weekly_sales'].shift(periods = 4)
weekly_sales_lags['sales_lag_q'] = weekly_sales_lags['weekly_sales'].shift(periods = 13)
weekly_sales_lags['sales_lag_y'] = weekly_sales_lags['weekly_sales'].shift(periods = 52)

plt.figure(figsize=(8, 8))
mask = np.zeros_like(weekly_sales_lags.corr())
mask[np.triu_indices_from(mask)] = True

sns.heatmap(weekly_sales_lags.corr(),
            mask=mask,
            annot=True,
            cmap='coolwarm',
            vmin=-1, 
            vmax=1);
```

**3b. What can you say about the series based on these autocorrelations?**

We can see above that autoorrelation peaks at 1 and 4 week lags. We also see strong annual correlations.

## Modeling

**5. Train/Test Split**  

Modeling with time series data presents unique challenges because our observations are not independent of one another. 

So, in splitting the data into `training` and `test` sets, **we cannot split randomly**. Use the first *75%* of observations as training data and the following 25% for test data.

```python
train, test = train_test_split(weekly_sales['weekly_sales'],
                                   test_size = 0.25,
                                   shuffle = False)
```

#### 6. Autoregressive Model

We will start by building a simple autoregressive model first: the `AR(1)` model. (*Note that this is the same thing as an ARIMA(1,0,0) model.*)

**6a. Instantiate and fit an `AR(1)` model.**

```python
arima = ARIMA(endog = train.astype(float),
              order = (1, 0, 0))
ar1 = arima.fit()
```

```python
pred_train = arima.predict(params = ar1.params, 
                         start = train.index[0], 
                         end = train.index[-1])
pred_test = arima.predict(params = ar1.params, 
                         start = test.index[0], 
                         end = test.index[-1])
```

**6b. Calculate the Mean Squared Error of the `AR(1)` on your training set and your test set.**


```python
train_rmse = np.sqrt(mean_squared_error(train, pred_train))
test_rsme = np.sqrt(mean_squared_error(test, pred_test))
print(train_rmse, test_rsme)
```

#### 9. Stationary Data

Remember that one assumption to using the ARIMA model is that our data are **stationary**. This is a necessary assumption for our model, and it is important for us to check for it whenever we fit a time series model.

**9a. Use the Augmented Dickey-Fuller test to estimate whether or not your data are stationary.**

```python
adfuller(weekly_sales)
```

```python
# Code written by Joseph Nelson.
def interpret_dftest(df):
    dftest = adfuller(df)
    dfoutput = pd.Series(dftest[0:2], index=['Test Statistic','p-value'])
    return dfoutput
interpret_dftest(weekly_sales)
```

**9b. Interpret the results of the Augmented Dickey-Fuller test.**

We have sufficient evidence to reject H_0 at 99% significance and conclude our data is stationary.

#### 12.  SARIMAX Model
We have looked at autocorrelation of neighboring values (i.e. comparing $Y_t$ to $Y_{t-1}$ or $Y_t$ to $Y_{t-4}$). However, what about seasonal affects on our data? How might the sales today be related to sales 6 months ago or a year ago?

Let's build a SARIMAX model!

**12a. Using `PACF` and `ACF` plots with appropriate intervals (think months in a year) try to build the best SARIMAX model you can in setting the `seasonal_order = (P, D, Q, S)` and using `order = (2, 0, 2)`, fit and evaluate the model as we have done in the prior models**.

```python
# Decompose time series into trend, seasonal, and residual components.
decomp = seasonal_decompose(weekly_sales)

plt.figure(figsize=(16,12))
decomp.plot();
```

```python
# Taken form Raphael Bubolz Larrosa at TDS
# https://towardsdatascience.com/how-to-forecast-sales-with-python-using-sarima-model-ba600992fa7d
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 52) for x in list(itertools.product(p, d, q))]

for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = SARIMAX(train.astype(float).dropna(),
                          order=param,
                          seasonal_order=param_seasonal)
            results = mod.fit()
            print('ARIMA{}x{}52 - AIC:{}'.format(param,param_seasonal,results.aic))
        except: 
            continue
```

Our AIC search suggests (0,1,1)(1,1,1,52) provides the best option. We also see a wide variety of close AIC criteria for low order models with seasonal affects.

```python

sarima = SARIMAX(endog = train.astype(float).dropna(),
                 order = (0, 1, 1), 
                 seasonal_order = (1, 1, 1, 52))
sar = sarima.fit()
print(sar.summary())
```

```python
sar.plot_diagnostics(figsize=(18, 8))
plt.show()
```

```python
# Visual overview of performance on training dataset
pred = sar.get_prediction(start=train.index[0], dynamic=False)
pred_ci = pred.conf_int()
ax = train.plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7, figsize=(14, 4))
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('Retail_sold')
plt.legend()
plt.show()
```

```python
# Visualize test performance
pred = sar.predict(test.index[0], test.index[-1], typ='levels')
ax = test.plot(label='observed')
pred.plot(ax=ax, label='SARIMA', alpha=.7, figsize=(14, 4))
ax.set_xlabel('Date')
ax.set_ylabel('Retail_sold')
plt.legend()
plt.show()
```

```python
np.sqrt(((test - pred)**2).mean())
```

**12b. How did your best SARIMAX model perform?**

We got down to a root mean square error of $107,660 which represents an average error of 10\% of actual sales. This is actually higher than our ARMA(2,2) model. 

## Repeat above with better params

We choose to disregard our AIC search and choose to use AR 2 parameters from above which showed the lowest error rate without seasonal effects.

```python
p = 2
d = q = 0
pdq = [(p, d, q)]
seasonal_pdq = [(x[0], x[1], x[2], 52) for x in list(itertools.product(range(2), range(2), range(2)))]

for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = SARIMAX(train.astype(float).dropna(),
                          order=param,
                          seasonal_order=param_seasonal)
            results = mod.fit()
            print('ARIMA{}x{}52 - AIC:{}'.format(param,param_seasonal,results.aic))
        except: 
            continue
```

```python
sarima = SARIMAX(endog = train.astype(float).dropna(),
                 order = (2, 0, 0), 
                 seasonal_order = (1, 1, 1, 52))
sar = sarima.fit()
print(sar.summary())
```

```python
pred = sar.predict(test.index[0], test.index[-1], typ='levels')
np.sqrt(((test - pred)**2).mean())
```

We reduced our root mean square error by 30% by ignoring our AIC search and choosing to build off our simpler AR2 model from above.

### 14. BONUS: Rolling Predictions

**Try testing the performance of your model using rolling predictions. For each testing observation $Y_t$ you should train your model using data up to $Y_{t-1}$. This allows us to make meaningful predictions for all testing observations** 

```python
train, test = train_test_split(weekly_sales['weekly_sales'],
                                   test_size = 0.25,
                                   shuffle = False)
```

```python
pred_roll = list()
for date in test.index:
    
    sar_model = SARIMAX(endog = train.dropna(),
                 order = (2, 0, 0), 
                 seasonal_order = (1, 1, 1, 52))
    sar_fit = sar_model.fit()
    yhat = sar_fit.predict(start = date, end = date, typ='levels')
    pred_roll.append(yhat[0])
    train = weekly_sales.loc[weekly_sales.index <= date,'weekly_sales']
```

```python
pred_roll[0]
```

```python
np.sqrt(((test - pred_roll)**2).mean())
```

Updating our model as we move forward in time reduced our root mean square error by 2 percent.
